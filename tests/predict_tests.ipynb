```python
# PredUtils.py
import numpy as np
from genet.predict.PredUtils import preprocess_seq, get_seq

def preprocess_seq(data, seq_length):
    """
    Preprocess a list of DNA sequences into a one-hot encoded numpy array.

    Args:
        data (list): A list of DNA sequences.
        seq_length (int): The length of the DNA sequences.

    Returns:
        seq_onehot (numpy.ndarray): A one-hot encoded numpy array of shape
            (len(data), 1, seq_length, 4).
    """
    seq_onehot = np.zeros((len(data), 1, seq_length, 4))
    for i, seq in enumerate(data):
        for j, nucleotide in enumerate(seq):
            if nucleotide == 'A':
                seq_onehot[i, 0, j, 0] = 1
            elif nucleotide == 'C':
                seq_onehot[i, 0, j, 1] = 1
            elif nucleotide == 'G':
                seq_onehot[i, 0, j, 2] = 1
            elif nucleotide == 'T':
                seq_onehot[i, 0, j, 3] = 1
    return seq_onehot

# Define test data
data = ['ATCG', 'CGTA']
seq_length = 4

# Preprocess the test data
seq_onehot = preprocess_seq(data, seq_length)

# Print the output
print(seq_onehot)
 
assert seq_onehot.shape == (2, 1, 4, 4)
assert np.all(seq_onehot.sum(axis=3) == 1)
assert np.all(seq_onehot.sum(axis=2) == 1)
```
